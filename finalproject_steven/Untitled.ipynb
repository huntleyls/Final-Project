```python
benign_domain_df = pd.read_csv("benign_domain_list.csv", header=None)
benign_domain_df = benign_domain_df.rename(columns={0: "domain"}, errors="raise")
malicious_url_df = pd.read_csv("Malicious_URls.csv")
malicious_url_df
```

```python
domainlist = []
tldlist = []
for url in malicious_url_df['url']:
    ext = tldextract.extract(url)
    domain = ext.domain
    tld = ext.suffix
    domainlist.append(domain)
    tldlist.append(tld)
    
malicious_url_df['domain'] = domainlist
malicious_url_df['tld'] = tldlist
malicious_url_df
```
